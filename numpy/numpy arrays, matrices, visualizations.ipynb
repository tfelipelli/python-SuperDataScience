# defining a function

```python
def myplot(data, playerlist = Players):
    
    colors = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red","KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta"}
    
    markers = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^","KevinDurant":"D","DerrickRose":"s","DwayneWade":"o"}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=colors[name], ls="--", marker=markers[name], ms=5, label=name)
    
    plt.legend(loc="upper left", bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation = "vertical")

    plt.show()
```

```python
myplot(Points, ["KobeBryant", "LeBronJames", "DerrickRose", "JoeJohnson"])
```

```python
# data insights
myplot(Salary)
myplot(Salary / Games)
myplot(Salary / FieldGoals)
```

```python
# in-game metrics
myplot(MinutesPlayed)
myplot(Points)
```

```python
# in-game metrics normalized -> taking injuries out of the equation
myplot(FieldGoals / Games)
myplot(FieldGoals / FieldGoalAttempts)
myplot(FieldGoalAttempts / Games)
myplot(Points / Games)
```

```python
# interesting observations
myplot(MinutesPlayed / Games)
myplot(Games)
```

```python
myplot(FieldGoals / MinutesPlayed)
```

```python
# player style
myplot(Points / FieldGoals)
```
