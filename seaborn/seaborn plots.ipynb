```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
plt.rcParams["figure.figsize"] = 10, 8 # to fix size of plot
warnings.filterwarnings('ignore')
```

```python
stats = pd.read_csv("/home/thais/Documents/code/learning/DemographicData.csv")
# renaming columns
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
```

```python
stats.head()
```

```python
# distribution
vis1 = sns.distplot(stats["InternetUsers"])
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
```

```python
vis2 = sns.distplot(stats["BirthRate"], bins=30)
```

```python
# boxplots
vis3 = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

```python
# seaborn gallery
# https://seaborn.pydata.org/examples/index.html
```

-------------------------------

```python
# linear (regression) model 
vis4 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False)
```

```python
vis4 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=7, aspect=2)
```

-------------------------------

```python
# KEYWORD ARGUMENTS IN PYTHON <> KWS

# to go to definition of method -> F12
# seaborn relies on matplotlib
# https://seaborn.pydata.org/generated/seaborn.lmplot.html
# {scatter,line}_kwsdictionaries
#   Additional keyword arguments to pass to plt.scatter and plt.plot

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter


# marker size
vis5 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", height=7, aspect=2, scatter_kws={"s":150})
```
